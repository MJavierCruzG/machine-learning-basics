### Yahoo finance

```python
start = datetime.datetime(2014, 1, 1)
end = datetime.datetime(2016, 7, 15)
yahoo_df = data.DataReader("F", 'iex', start, end)
yahoo_df.plot()
```

### FRED
source: http://quant-econ.net/py/pandas.html

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2016, 7, 15)
unrate_df = data.DataReader('UNRATE', 'fred', start, end)
unrate_df.plot()
```

### Google

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2016, 7, 15)
google_df = data.DataReader("F", 'robinhood', start, end)
google_df.plot()
```
